# Diff eb(left) vs spack(right)

```python
# Builder map for branch "spack-builder-tollerances-round2" revision "2552abad9f91a5d93a046214bb6520d44507a8f1"
# test suite app https://octopus-code.org/testsuite/commits/1636
# eb pipeline https://gitlab.com/octopus-code/octopus/-/pipelines/1153321045
# spack pipeline https://octopus-code.org/buildbot/#/builders/137/builds/28

# builder_map = {builder:{easybuild:eb_url,spack:spack_url}}
example_map = {
    "foss2022a_mpi": {
        "easybuild": "https://www.octopus-code.org/buildbot/#/builders/108/builds/207",
        "spack": "https://octopus-code.org/buildbot/#/builders/102/builds/48",
    },
    "foss2022a_mpi_omp": {
        "easybuild": "https://www.octopus-code.org/buildbot/#/builders/53/builds/1236",
        "spack": "https://octopus-code.org/buildbot/#/builders/138/builds/31",
    },
    # "foss2022a_mpi_debug": {
    #     "easybuild": "https://www.octopus-code.org/buildbot/#/builders/161/builds/184",
    #     "spack": "https://octopus-code.org/buildbot/#/builders/152/builds/20",
    # },
    # # "foss2022a_cuda_mpi_omp": {
    #     "easybuild": "https://www.octopus-code.org/buildbot/#/builders/147/builds/240",
    #     "spack": "https://octopus-code.org/buildbot/#/builders/159/builds/14",
    # },
}
```

```python
for builder in example_map:
    left_url = example_map[builder]["easybuild"]
    right_url = example_map[builder]["spack"]
    headding = Markdown(f"""
## {builder=}
### {left_url=}
### {right_url=}
""")
    rprint(headding)
    try:
        left_config_h = get_config_h(left_url)
        right_config_h = get_config_h(right_url)
        diff = get_diff(
            left_config_h, right_config_h, f"eb_{builder}", f"spack_{builder}"
        )
        rprint(diff)
    except:
        rprint(Markdown("### failed to compile results"))
```
