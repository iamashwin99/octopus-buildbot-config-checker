```python
from config_diff import get_config_h, get_diff
from rich.markdown import Markdown
from rich import print as rprint
```

```python
example_map = {
        "foss-2022a_cuda_mpi_omp": {
        "easybuild": "https://www.octopus-code.org/buildbot/#/builders/147/builds/289",
        "spack": "https://www.octopus-code.org/buildbot/#/builders/159/builds/24",
    },
}


for builder in example_map:
    left_url = example_map[builder]["easybuild"]
    right_url = example_map[builder]["spack"]
    headding = Markdown(f"""
## {builder=}
### {left_url=}
### {right_url=}
""")
    rprint(headding)
    try:
        left_config_h = get_config_h(left_url)
        right_config_h = get_config_h(right_url)
        diff = get_diff(
            left_config_h, right_config_h, f"eb_{builder}", f"spack_{builder}"
        )
        rprint(diff)
    except:
        rprint(Markdown("### failed to compile results"))
```
